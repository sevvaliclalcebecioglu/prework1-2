## _Ek Veri Çerçevesi İşlemleri_

```python
# CSV'leri oku ve sütun isimlerini büyük harfe çevir
df = pd.read_csv("client_list.csv").rename(columns=str.upper)
ids = pd.read_csv("customer_id_list.csv")
```

```python
# Müşteri listelerini birleştir (merge)
clients = df.merge(ids, how="left", on=["LAST_NAME","FIRST_NAME","GENDER","AGE"])  # LEFT JOIN ile ID bilgilerini ekler
```

```python
# Satış verilerini oku ve müşterilerle birleştir
sales = pd.read_csv("purchase_log.csv")
detailed_sales = clients.merge(sales, how="inner", on="CUSTOMER_ID")  # INNER JOIN ile sadece satın alma yapanları alır
```

```python
# Müşterilerin toplam harcamasını hesapla
spenders = detailed_sales.groupby(
    ["CUSTOMER_ID","FIRST_NAME","LAST_NAME","GENDER","AGE"], as_index=False
)["PRICE"].sum()  # Her müşteri için toplam harcama
```

```python
# En çok harcama yapan müşteriyi bul
top_spender = spenders.loc[spenders.PRICE.idxmax()]  # Maksimum PRICE değerine sahip satırı seçer
```

```python
# BONUS: En çok harcayan müşterinin satın aldığı ürünleri listele
top_spender_items = sales.loc[sales.CUSTOMER_ID==top_spender.CUSTOMER_ID, "ITEM_DESCRIPTION"]  # Toplam harcayanın ürünleri
```

## _Dosya Yazma_

```python
# detailed_sales veri çerçevesini CSV olarak kaydet (index olmadan)
detailed_sales.to_csv("df_out.csv", index=False)  # CSV dosyası oluşturur
```

```python
# detailed_sales veri çerçevesini pickle formatında kaydet
detailed_sales.to_pickle("df_out.p")  # Pickle ile hızlı yüklenebilir veri dosyası oluşturur
```
