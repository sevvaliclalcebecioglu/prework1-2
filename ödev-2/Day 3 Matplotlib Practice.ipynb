# _Matplotlib - Solutions_

_Bar Chart: Ay isimlerindeki harf sayılarını görselleştirir, kenar çizgisiz ve yeşil renklidir._

```python
import matplotlib.pyplot as plt
from numpy import arange, array

# Veriler
x = arange(1, 13)
y = array([7, 8, 5, 5, 3, 4, 4, 6, 9, 7, 8, 8])

# Bar grafiği
plt.bar(x, y, width=0.25, color='green', edgecolor='none')  # Yeşil barlar, kenar çizgisi yok
plt.xlim(1, 13)  # X eksenini sabitle
plt.ylim(0, 10)  # Y eksenini sabitle
plt.xlabel("month of the year")  # X ekseni etiketi
plt.ylabel("letters in month name")  # Y ekseni etiketi
plt.title("Leters in the Name of Each Month")  # Başlık
plt.show()  # Grafiği göster
```

_Line & Scatter Plot: 2*cos(x) fonksiyonunu hem kesik yeşil çizgi hem de büyük mavi noktalar ile gösterir, grafiği "cos.png" olarak kaydeder._

```python
from numpy import arange, cos, pi

# Veriler
x = arange(0.0, 2 * pi, pi / 16.0)
y = 2 * cos(x)

plt.figure(figsize=(8,6))

# Line plot
plt.plot(x, y, linewidth=3, color='green', linestyle='--', label='approx')  # Yeşil kesik çizgi, label 'approx'

# Scatter plot
plt.scatter(x, y, s=1000, edgecolor='none', color='blue', label='discrete')  # Büyük mavi noktalar

# Etiketler ve başlık
plt.xlabel("x")
plt.ylabel("Y")
plt.title("2 * cos(x)")
plt.legend()

# Grafiği kaydet
plt.savefig("cos.png")
plt.show()
```

_Multiplot (Histogram + Scatter): Yan yana subplot; solda histogram, sağda rastgele renkli scatter plot, eksen limitleri ve grid ile görselleştirilmiş._

```python
from numpy import random
import matplotlib.pyplot as plt

# Veriler
mu = 10500
sigma = 250
values = mu + sigma * random.randn(10000)
x = mu + sigma * random.randn(100)
y = mu + sigma * random.randn(100)

# 2 kolonlu subplot
fig, axs = plt.subplots(1, 2, figsize=(14,6))

# Sol: Histogram
axs[0].hist(values, bins=25, facecolor='green', edgecolor='none')  # Yeşil histogram
axs[0].set_ylabel("count of values")  # Y ekseni etiketi

# Sağ: Scatter plot
colors = random.rand(100)
axs[1].scatter(x, y, s=300, alpha=0.25, edgecolor='none', c=colors)  # Saydam ve rastgele renkli noktalar
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].set_xlim(10000, 11000)  # X ekseni limiti
axs[1].set_ylim(10000, 11000)  # Y ekseni limiti
axs[1].grid(True)  # Grid ekle

plt.show()
```
